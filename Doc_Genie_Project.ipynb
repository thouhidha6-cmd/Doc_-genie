```python
import ast
import gradio as gr
from reportlab.pdfgen import canvas
```

```python
def extract_function_signature(code):
    tree = ast.parse(code)

    for node in ast.walk(tree):
        if isinstance(node, ast.FunctionDef):
            name = node.name
            params = []

            for arg in node.args.args:
                params.append(arg.arg)

            return name, params
```

```python
def generate_docstring(code):
    name, params = extract_function_signature(code)

    doc = f"{name} function\n\nParameters:\n"

    for p in params:
        doc += f"- {p}\n"

    return doc
```

```python
def generate(code):
    return generate_docstring(code)

interface = gr.Interface(
    fn=generate,
    inputs=gr.Textbox(lines=10, label="Paste Python Function"),
    outputs="text",
    title="Doc Genie - Python Documentation Generator"
)

interface.launch()
```
